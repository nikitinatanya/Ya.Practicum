## Введение

Отдел маркетинговой аналитики Яндекс.Афиши *анализирует расходы для последующей оптимизации и перераспределения маркетингового бюджета в пользу выгодных источников трафика*.
В наличии данные Яндекс.Афиши с июня 2017 по конец мая 2018 года:
- лог сервера с данными о посещениях сайта Яндекс.Афиши,
- выгрузка всех заказов за этот период,
- статистика рекламных расходов.  

**Задачи исследования:**
- изучить, как клиенты пользуются сервисом,
- понять, когда делают первые покупки на сайте,
- посчитать, сколько денег приносит компании каждый клиент,
- определить, когда расходы на привлечение клиента окупаются.  

Исследование включает в себя следующие **этапы**:
1. Изучение общей информации.
2. Подготовка данных.
3. Расчет метрик и визуализация:
    - продуктовые,
    - метрики электронной коммерции,
    - маркетинговые.
4. Общий вывод.


**Описание данных**  

 Файл visits_log.csv хранит лог сервера с информацией о посещениях сайта, orders_log.csv — информацию о заказах, а costs.csv — информацию о расходах на маркетинг.

#### Коэффициент удержания (Retention Rate)

Для расчета коэффициента удержания сформируем когорты. Определим событие и период. Возьмём дату, когда пользователь впервые зашел на сайт Афиша.Яндекс.  
Эти данные запишем в объект Series first_visit_month. Объединим его с исходным датафреймом visits.

```python
first_visit_date = visits.groupby(['uid'])['start_ts'].min()
first_visit_date.name = 'first_visit_date'
visits = visits.join(first_visit_date, on ='uid')
visits.head()
```

Выделим месяц из first_visit_month и start_ts (сформировав новый столбец visit_month).

```python
visits['first_visit_month'] = visits['first_visit_date'].astype('datetime64[M]')
visits['visit_month'] = visits['start_ts'].astype('datetime64[M]')
visits.head()
```

Сформируем когорты с периодом 1 месяц. Когортой станет столбец first_visit_month — месяц, в котором пользователь впервые зашел на сайт.

```python
visits['cohort_lifetime'] = visits['visit_month'] - visits['first_visit_month']
visits['cohort_lifetime'] = visits['cohort_lifetime'] / np.timedelta64(1,'M')
visits['cohort_lifetime'] = visits['cohort_lifetime'].round().astype('int')
visits.head()
```

Сгруппируем данные по когорте и lifetime. Посчитаем для каждой когорты количество активных пользователей на определённый «месяц жизни».

```python
cohorts = visits.groupby(['first_visit_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index()
cohorts.head()
```

Найдём исходное количество пользователей в когорте. 

```python
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][
    ['first_visit_month', 'uid']
]
initial_users_count.head()
```

```python
# переименование столбца uid
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'}) 
```

Объединим данные по когортам с исходным количеством пользователей в когорте.

```python
cohorts = cohorts.merge(initial_users_count,on='first_visit_month')
cohorts.head()
```

Рассчитаем Retention Rate. Разделим количество активных пользователей в каждую из недель на исходное число пользователей в когорте

####  Lifetime value покупателя (LTV)

Расчет LTV на покупателя производится по месячным когортам и включает в себя:  
- найти дату первой покупки каждого покупателя;
- вычислить число новых покупателей за каждую дату;
- добавить дату первой покупки в таблицу с заказами;
- сгруппировать таблицу и посчитать выручку;  
- посчитать валовую прибыль;  
- рассчитать возраст когорты;
- найти LTV.

Первые три шага были выполнены в рамках анализа среднего количества покупок на покупателя за 6 месяцев.

Сгруппируем таблицу заказов по месяцу первой покупки и месяцу каждого заказа и сложим выручку.

```python
cohorts_orders = (
    orders_new.groupby(['first_order_month', 'order_month'])
    .agg({'revenue': 'sum'})
    .reset_index()
)
cohorts_orders.head()
```

Добавим в таблицу cohorts_orders данные о том, сколько людей первый раз совершили покупку в каждый месяц.

```python
report = pd.merge(cohort_sizes, cohorts_orders, on='first_order_month')
report.head()
```

LTV считают не из выручки, а из валовой прибыли. Значит, нужно найти валовую прибыль, умножив выручку на маржинальность.

```python
margin_rate = 1.0

report['gp'] = report['revenue'] * margin_rate
report.head()
```

LTV — относительный показатель. Значит его удобнее смотреть по «возрастным» когортам, где в столбцах — не месяц заказа, а возраст (age) когорты.

```python
report['age'] = (
    report['order_month'] - report['first_order_month']
) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')
report.head()
```

Найдём LTV. Разделим валовую прибыль когорт за каждый месяц на общее число пользователей в каждой когорте. Для наглядности визуализируем когорты в виде сводной таблицы.

```python
report['ltv'] = report['gp'] / report['n_buyers']

output = report.pivot_table(
    index='first_order_month', 
    columns='age', 
    values='ltv', 
    aggfunc='mean'
).round(2)

output.index = output.index.strftime("%m-%Y")
output = output.cumsum(axis=1)
output.fillna('')
```

Рассчитаем средний LTV по когортам за 6 месяцев. В расчёт включены когорты, «прожившие» не менее 6 месяцев.

```python
output_age_6 = output.loc[output[5] > 0]
output_age_6.fillna('')
```

```python
sns.set(style='white')
plt.figure(figsize=(15, 8))
plt.title('Когорты: LTV', size=14)
ax=sns.heatmap(output_age_6, annot=True, fmt='.2f', linewidths=1, linecolor='gray')
ax.set_ylabel('Когорта')
ax.set_xlabel('Месяц жизни когорты')
plt.yticks(rotation = 0)
plt.show()
```

```python
print('Средний LTV на 6 месяц жизни когорты: {:.2f} у.е.'.format(output_age_6[5].mean()))
```

Самая прибыльная когорта 09/2017: в среднем каждый покупатель принес 13,43 у.е. валовой прибыли за 9 месяцев жизни. Когорта-долгожитель 06/2017: за год в среднем каждый покупатель принес 11,88 у.е. валовой прибыли.

#### Return on marketing investments (ROMI)

Расчет ROMI по когортам в разрезе источников осуществляется по следующему алгоритму:  
- определить первый источник источник трафика (сделано ранее first_source);  
- найти месяц первой покупки, месяц покупки (сделано ранее orders_new_first_source);  
- объединить first_source и orders_new_first_source;  
- рассчитать количество новых покупателей по каждому источнику трафика и месяцу первой покупки;  
- сгруппировать таблицу заказов по источнику трафика, месяцу первой покупки и месяцу покупки и посчитать выручку;  
- добавить в таблицу cohorts_orders_source данные о том, сколько пользователей первый раз совершили покупку в каждый месяц (объединить cohorts_orders_source и cohort_sizes_source в новый датафрейм report_source);  
- рассчитать валовую прибыль;  
- рассчитать возраст когорты;  
- определить затраты на рекламу по источникам трафика и месяцу (ad_costs_source);  
- дополнить таблицу report_source затратами на рекламу (объединить report_source и ad_costs_source);
- рассчитать CAC;  
- рассчитать ROMI;  
- сформировать срезы по каждому источнику и построить heatmap.


```python
# таблица orders_new -  с месяцем заказа, первым месяцем заказа
orders_new.head()
```

```python
orders_new_first_source['source_id'].unique()
```

```python
# количество покупателей по источникам и месяцу первой покупки
cohort_sizes_source = (
    orders_new_first_source.groupby(['source_id', 'first_order_month'])
    .agg({'uid': 'nunique'})
    .reset_index()
)
cohort_sizes_source.columns = ('source_id', 'first_order_month','n_buyers')
cohort_sizes_source.head()
```

```python
# расчет выручки в разрезе источников трафика, месяца первой покупки и месяца покупки
cohorts_orders_source = (
    orders_new_first_source.groupby(['source_id', 'first_order_month', 'order_month'])
    .agg({'revenue': 'sum'})
    .reset_index()
)
cohorts_orders_source.head()
```

```python
#report_source['revenue'].sum()
```

```python
# валовая прибыль
margin_rate = 1.0

report_source['gp'] = report_source['revenue'] * margin_rate
report_source.head()
```

```python
# расчет LTV
report_source['ltv'] = report_source['gp'] / report_source['n_buyers']
```

```python
# возраст когорты
report_source['age'] = (
    report_source['order_month'] - report_source['first_order_month']
) / np.timedelta64(1, 'M')
report_source['age'] = report_source['age'].round().astype('int')
report_source.head()
```

```python
# маркетинговые расходы в разрезе источников трафика и месяца затрат
ad_costs_source_month = ad_costs.groupby(['source_id', 'cost_month']).agg({'costs' : 'sum'}).reset_index()
ad_costs_source_month.head()
```

```python
# расчет CAC
report_source_new = pd.merge(
    report_source, ad_costs_source_month, \
    left_on=['source_id', 'first_order_month'], right_on=['source_id','cost_month']
)
report_source_new['cac'] = report_source_new['costs'] / report_source_new['n_buyers']
report_source_new.head()
```

```python
# расчет ROMI
report_source_new['romi'] = report_source_new['ltv'] / report_source_new['cac']
report_source_new.head()
```
