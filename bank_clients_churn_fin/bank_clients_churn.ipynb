### Анализ  клиентов по признаку оттока

Посмотрим на средние значения признаков в двух группах — тех, кто ушел в отток и тех, кто остался

```python
bank_churn = bank['churn'].value_counts().reset_index()
bank_churn = bank_churn.rename(columns={'index' : 'gone/stayed', 'churn':'count'})
bank_churn['ratio'] = bank_churn['count'] / bank_churn['count'].sum()
display(bank_churn.style.format({'ratio':"{0:.1%}"}))
```

Общее количество пользователей - 10000, из них 2037 (20.4%) перестали пользоваться услугами банка.

```python
fig, axs = plt.subplots(figsize=(15, 8))
#explode = [0.07, 0.07, 0.07, 0, 0, 0, 0, 0, 0]
bank_churn['count'].plot(
    kind='pie',
    #explode=explode,
    startangle=45,
    labels = ['Пользователь остался', 'Пользователь ушел'],
    #labeldistance = 1.1,
    autopct='%1.1f%%',
    shadow=True, 
    counterclock = False,
    #frame = True,
    #rotatelabels = True
)
axs.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.xlabel('Признак оттока')
plt.ylabel('Доля')
plt.figtext(.5,.9,'Структура пользователей по признаку оттока', fontsize=27, ha='center')
plt.show()
```

```python
columns_without_churn = ['score', 'city', 'gender', 'age', 'objects', 'balance',
       'products', 'credit_card', 'loyalty', 'estimated_salary']
```

```python
bank_churn_mean = bank.drop('user_id', axis = 1).groupby('churn').agg({'mean'}).T\
   .rename(columns={0 : 'client_stayed', 1 : 'client_gone'})
bank_churn_mean['ratio'] = bank_churn_mean['client_gone'] / bank_churn_mean['client_stayed'] - 1
bank_churn_mean = bank_churn_mean.sort_values(by ='ratio', ascending = False)

display(bank_churn_mean.style.format({'ratio':"{0:.0%}"}))
```

Построим столбчатые гистограммы и распределения признаков для тех, кто ушёл (отток) и тех, кто остался (не попали в отток).


```python
for column in columns_without_churn:
    fig, axes = plt.subplots(1, 2, figsize = (15,4))
    plt.figtext(.5,.9,'Гистограмма ' + column, fontsize=27, ha='center')

    bank.query('churn == 0').plot(y=column, kind='hist',linewidth=2, ax = axes[0], legend = False, grid = True).set(ylabel = 'Количество клиентов, чел.', xlabel = 'Остался')
    bank.query('churn == 1').plot(y=column, kind='hist',linewidth=2, ax = axes[1], legend = False, grid = True).set(ylabel = 'Количество клиентов, чел.', xlabel = 'Попал в отток')
    plt.show()
```

#### Вывод

Анализ средних значений признаков и их распределения в разрезе групп пользователей "остался" и "попал в отток".  

*Высокое различие типичных значений групп:*
- **гендерная принадлежность:** в структуре оттока в сравнении с со структурой остающихся больше женщин (на 35%);  
- **баланс:** в отток попадают клиенты с более высоким дебетовым остатком на счету (разница в средних значениях групп - 25%). ## Модель оттока клиентов

Построим бинарной классификации, где целевой признак — факт оттока клиента.   


Признаков с высокой корреляцией нет. Есть формальный признак, не влияющий на отток - `user_id` - удалим его.

```python
bank_mod = bank.drop('user_id', axis = 1)
```

Разделим наши данные на признаки (матрица X) и целевую переменную (y).

```python
X = bank_mod.drop('churn', axis = 1)
y = bank_mod['churn']
```

Разделим данные на обучающую и валидационную выборку

```python
#разделим модель на обучающую и валидационную выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# создадим объект класса StandardScaler и применим его к обучающей выборке
scaler = StandardScaler()
# обучаем scaler и одновременно трансформируем матрицу для обучающей выборки
X_train_st = scaler.fit_transform(
    X_train
)  
# применяем стандартизацию к матрице признаков для тестовой выборки
X_test_st = scaler.transform(X_test)
```

Обучим модель на train-выборке алгоритмом логистической регрессии.

```python
# зададим алгоритм для модели логистической регрессии
lr_model = LogisticRegression(solver = 'lbfgs') 
# обучим модель
lr_model.fit(X_train_st, y_train)
# воспользуемся уже обученной моделью, чтобы сделать прогнозы
lr_predictions = lr_model.predict(X_test_st)
lr_probabilities = lr_model.predict_proba(X_test_st)[:,1]
```

Оценим метрики accuracy (доля правильных ответов), precision (точность) и recall (полнота).

```python
'''
Функция, которая рассчитывает метрики
'''

def print_all_metrics(y_true, y_pred, y_proba, title='Метрики классификации'):
    print(title)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
    print('\tF1: {:.2f}'.format(f1_score(y_true, y_pred)))
    #print('\tROC_AUC: {:.2f}'.format(roc_auc_score(y_true, y_proba)))
```

```python
print_all_metrics(y_test, lr_predictions, lr_probabilities, 'Метрики для модели логистической регрессии:')
```

Оценим количественное влияние признаков на целевую переменную - рассмотрим коэффициенты логистической регрессии.

```python
lr_featureImportances=pd.DataFrame(data=lr_model.coef_,columns=X.columns).T
lr_featureImportances=lr_featureImportances.apply(lambda x: abs(x))
lr_featureImportances.sort_values([0],ascending=False,inplace=True)
lr_featureImportances.columns = ['lr_coef_abs']

lr_featureImportances
```

В рамках данного исследования для оценки количественного влияния признаков на отток мы будем использовать модель логистической регрессии. Выявленные линейные зависимости однозначно интерпретируются и точно сработают на практике. 
Вместе с тем для сравнения посмотрим модель, построенную алгоритмом случайного леса. Выявленные нелинейные зависимости можно использовать впоследствие для более глубокой проработки клиентской базы и запуске тестовых маркетинговых кампаний.

```python
# зададим алгоритм для модели на основе алгоритма случайного леса
rf_model = RandomForestClassifier(n_estimators = 100,random_state = 0)
# обучим модель случайного леса
rf_model.fit(X_train_st, y_train)
# воспользуемся уже обученной моделью, чтобы сделать прогнозы
rf_predictions = rf_model.predict(X_test_st)
rf_probabilities = rf_model.predict_proba(X_test_st)[:,1]
```

```python
print_all_metrics(y_test, rf_predictions, rf_probabilities, 'Метрики для модели случайного леса:')
```

```python
rf_featureImportances=pd.DataFrame(data=rf_model.feature_importances_, index=X.columns)
rf_featureImportances=rf_featureImportances.apply(lambda x: abs(x))
rf_featureImportances.sort_values([0],ascending=False,inplace=True)
rf_featureImportances.columns = ['rf_feat_abs']
rf_featureImportances
```

```python
featureImportances_join = lr_featureImportances.join(rf_featureImportances)
featureImportances_join
```

### Возраст клиента

Распределим клиентов на группы по возрасту:
- до 30 лет;
- 30-39 лет;
- 40-49 лет;
- 50-60 лет;
- старше 60.

```python
def age_group(row):
    '''
    Функция распределения клиентов на группы по возрасту
    '''
    age = row['age']
    if age < 30: return 'до 30'
    if age >= 30 and age < 40: return '30-39 лет'
    if age >= 40 and age < 50: return '40-49 лет'
    if age >= 50 and age <= 60: return '50-60 лет'
    else: return 'старше 60'
```

```python
bank['age_group'] = bank.apply(age_group, axis=1)
```

```python
#рассчитаем отношение клиентов в оттоке к общему количеству клиентов по возрастным группам
total_age = bank['age_group'].value_counts().to_frame().reset_index().rename(columns = {'index':'age_group','age_group':'total_clients'})
churn_age = bank.query('churn == 1')['age_group'].value_counts().to_frame().reset_index().rename(columns = {'index':'age_group','age_group':'churn_clients'})
age_count = total_age.merge(churn_age, how = 'left', on = 'age_group').sort_values('churn_clients', ascending = False)
age_count['ratio'] = age_count['churn_clients'] / age_count['total_clients']
age_count.style.format({'ratio':'{:.1%}'})
#age_count
```

```python
fig, axs = plt.subplots(figsize=(15, 8))
age_count['churn_clients'].plot(
    kind='pie',
    startangle=45,
    labels = age_count['age_group'].to_list(),
    #labeldistance = 1.1,
    autopct='%1.1f%%',
    shadow=True, 
    counterclock = False,
    #frame = True,
    #rotatelabels = True
)
axs.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.xlabel('Группа по возрасту')
plt.ylabel('Доля')
plt.figtext(.5,.9,'Структура оттока по возрастным группам', fontsize=27, ha='center')
plt.show()
```

## Проверка статистических гипотез

Проверим следущие гипотезы:
1. Доход между теми клиентами, которые ушли и теми, которые остались, различается;  
2. Средний возраст активных и неактивных клиентов различается.
    
`Проанализировать распределение признаков в целях выбора критерия проверки гипотез.  
Сформулировать нулевую и альтернативную гипотезы.  
Определить пороговое значение alpha.`  

### Доход между теми клиентами, которые ушли и теми, которые остались, различается

```python
# сформируем датафреймы по группам клиентов - клиент остался и клиент ушел
churn_0 = bank.query('churn == 0') # клиент остался
churn_1 = bank.query('churn == 1')# клиент ушел
```

```python
print('Средний доход группы "Клиент остался": {:.2f}'.format(churn_0['estimated_salary'].mean()))
print('Средний доход группы "Клиент ушел": {:.2f}'.format(churn_1['estimated_salary'].mean()))
print('Абсолютное отклонение доходов групп: {:.2f}'.format(churn_1['estimated_salary'].mean()-churn_0['estimated_salary'].mean()))
print('Относительное отклонение доходов групп: {:.2%}'.format(\
(churn_1['estimated_salary'].mean() - churn_0['estimated_salary'].mean()) / (churn_1['estimated_salary'].mean() + churn_0['estimated_salary'].mean())))
```

Проверим, является ли данная разница статистически значимой.

*Нулевая гипотеза*.  
Доход группы "Клиент остался" равен доходу группы "Клиент ушел".   

*Альтернативная гипотеза*.
Доход группы "Клиент остался" НЕ равен доходу группы "Клиент ушел".  

Проверим, как распределен показатель дохода (`estimated_salary`) в группах "Клиент остался" и "Клиент ушел".

```python
# распределение зарплаты по группам клиентов
ax = churn_0.plot(
    kind='hist',
    y='estimated_salary',    
    histtype='bar',
    #range=(1, 1700),
    bins=20,
    figsize = (15, 7),
    linewidth=2,
    #alpha=0.7,
    label='Клиент остался',
)

churn_1.plot(
    kind='hist',
    y='estimated_salary',
    histtype='bar',
    #range=(1, 1700),
    bins=20,
    figsize = (15, 7),
    title ='Распределение дохода по группам оттока',
    linewidth=2,
    alpha=0.8,
    label='Клиент ушел',
    ax=ax,
    grid=True,
    legend=True,
)


plt.xlabel('Доход, у.е.')
plt.ylabel('Количество клиентов, чел.')
plt.show()
```

Сравним дисперсии каждой группы по доходу.

```python
var_churn_0 = np.var(churn_0['estimated_salary'], ddof=1)
print('Дисперсия признака доход в группе "Клиент остался": {:.2f}'.format(var_churn_0))
var_churn_1 =  np.var(churn_1['estimated_salary'], ddof=1)
print('Дисперсия признака доход в группе "Клиент ушел": {:.2f}'.format(var_churn_1))
```

Используем статистический тест, чтобы подтвердить, что результаты, полученные из обоих распределений, не являются нормальным распределением.

```python
stat, p = normaltest(churn_0['estimated_salary'])
print('Statistics=%.3f, p-value=%.3f' % (stat, p))
alpha = 0.05
if p >= alpha:
    print('Принять гипотезу о нормальности')
else:
    print('Отклонить гипотезу о нормальности')
```

```python
stat, p = normaltest(churn_1['estimated_salary'])
print('Statistics=%.3f, p-value=%.3f' % (stat, p))
alpha = 0.05
if p >= alpha:
    print('Принять гипотезу о нормальности')
else:
    print('Отклонить гипотезу о нормальности')
```

Признак `estimated_salary` не распределен нормально. Для проверки гипотез будем использовать U-критерий Манна-Уитни. 

```python
alpha = 0.05

results = st.mannwhitneyu(churn_0['estimated_salary'], churn_1['estimated_salary'])

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя'
    )
```

#### Вывод  
Доход в группах "Клиент остался" и "Клиент ушел" примерно одинаковый. Разница не является статистически значимой.

Средний возраст группы "Клиент активный" равен среднему возрасту группы "Клиент неактивный".
