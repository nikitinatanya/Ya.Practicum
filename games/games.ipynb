## Введение

Интернет-магазине «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Необходимо выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании. Данные охватывают период до 2016 года.

**Цель** исследования: выявить закономерности, определяющие успешность компьютерной игры. Результаты  исследования - основание для планирования стратегии продаж в 2017 году.  

В ходе исследования рынка компьютерных игр перед нами стоят следующие **вопросы**:  

*1. Исследовательский анализ данных:*  
    - оценить, сколько игр выпускалось в разные годы; определить важны ли данные за все периоды;  
    - оценить, как менялись продажи по платформам; выбрать платформы с наибольшими суммарными продажами и построить распределение по годам; выявить характерный срок, за который появляются новые и исчезают старые платформы;  
    - на основании выводов на предыдущих шагах выбрать актуальный период и учитывать в анализе данные за этот период, при учёте, что эти данные помогут построить прогноз на 2017 год;  
    - определить какие платформы лидируют по продажам, растут или падают; выбрать несколько потенциально прибыльных платформ;  
    - построить график *«ящик с усами»* по глобальным продажам каждой игры и разбивкой по платформам; определить велика ли разница в продажах, в средних продажах на разных платформах;  
    - определить, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков; построить диаграмму рассеяния и посчитать корреляцию между отзывами и продажами;  
    - соотнести выводы с продажами игр на других платформах;  
    - оценить общее распределение игр по жанрам; оценить самые прибыльные жанры; определить выделяются ли жанры с высокими и низкими продажами.  

*2. Составление портрета пользователя каждого региона.*    
Определить для пользователя каждого региона (NA, EU, JP):
- самые популярные платформы (топ-5); описать различия в долях продаж;
- самые популярные жанры (топ-5); описать разницу;
- влияние рейтинга ESRB на продажи.  

*3. Проверка гипотез.*  
- средние пользовательские рейтинги платформ *Xbox One* и *PC* одинаковые;
- средние пользовательские рейтинги жанров *Action* и *Sports* разные.  

Исследование рынка компьютерных игр включает в себя следующие **этапы**:
1. Изучение общей информации.
2. Подготовка данных.
3. Исследовательский анализ данных.
4. Проверка гипотез.
5. Общий вывод.


**Описание данных**  

Name — название игры  
Platform — платформа  
Year_of_Release — год выпуска  
Genre — жанр игры  
NA_sales — продажи в Северной Америке (миллионы проданных копий)  
EU_sales — продажи в Европе (миллионы проданных копий)  
JP_sales — продажи в Японии (миллионы проданных копий)  
Other_sales — продажи в других странах (миллионы проданных копий)  
Critic_Score — оценка критиков (максимум 100)  
User_Score — оценка пользователей (максимум 10)  
Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

## Подготовка данных

В рамках данного этапа проанализируем корректность наименований столбцов таблицы, обработаем пропуски, изменим типы данных и дополним таблицу показателями.

### Обработка наименований столбцов

```python
# перевод наименования столбцов к нижнему регистру
data.columns = data.columns.str.lower()
data.columns
```

### Обработка пропусков

**Столбец name и столбец genre**

```python
data.loc[data['name'].isna(), :]
```

Количество пропусков в столбцах name и genre незначительно. Кроме того пропуски относятс к неактуальному для исследования периоду (1993 год). Удалим данные из таблицы.

```python
# удаление пропусков в столбце name
data.dropna(subset=['name'], inplace = True)
print('Количество пропусков в столбце name:', data['name'].isna().sum())
```

**Столбец year_of_release**

```python
print('Количество пропусков в столбце year_of_release:', data['year_of_release'].isna().sum())
print('Доля пропусков в столбце year_of_release:{:.2%}'.format(data['year_of_release'].isna().mean()))
```

Доля пропущенных значений в столбце year_of_release не превышает уровень существенности. Удалим строки с пропущенными значениями без ущерба для качества данных.

```python
# удаление пропусков в столбце year_of_release
data.dropna(subset=['year_of_release'], inplace = True)
print('Количество пропусков в столбце year_of_release:', data['year_of_release'].isna().sum())
```

```python
data['year_of_release'].sort_values(ascending = True).unique()
```

**Столбец critic_score**

```python
print('Количество пропусков в столбце critic_score:', data['critic_score'].isna().sum())
print('Доля пропусков в столбце critic_score:{:.2%}'.format(data['critic_score'].isna().mean()))
```

```python
data['critic_score'].sort_values(ascending = True).unique()
```

Количество пропусков в столбце critic_score существенно (51,5%). Отсутствие данных объективно, так как оценке критиков подвергаются не все игры, а наиболее популярные или как вариант имеющие аггресивную тактику продвижения на рынок.  
Пропуски оставим незаполненными.

**Столбец user_score**

```python
print('Количество пропусков в столбце user_score:', data['user_score'].isna().sum())
print('Доля пропусков в столбце user_score:{:.2%}'.format(data['user_score'].isna().mean()))
```

```python
data['user_score'].sort_values(ascending = True).unique()
```

В столбце user_score кроме числовых значений и nan есть значение tbd. В контексте компьютерных игр это аббревиатура расшифровывается To Be Determined, что означает, что рейтинг будет определен. То есть по сути это nan. Произведем замену.

```python
data.loc[data['user_score'] == 'tbd', 'user_score'] = np.nan
data['user_score'].sort_values(ascending = True).unique()
```

**Столбец rating**

```python
print('Количество пропусков в столбце rating:', data['rating'].isna().sum())
print('Доля пропусков в столбце rating:{:.2%}'.format(data['rating'].isna().mean()))
```

```python
data['rating'].sort_values(ascending = True).unique()
```

Количество пропусков в столбце rating существенно (40,6%). Отсутствие данных объективно, так как рейтинг носит региональный характер. Пропущенные значения заменим на "no_rating" для дальнейшего анализа зависимости продаж от рейтинга.

```python
data['rating'].fillna('no_rating', inplace=True)
```

В рейтинге есть маркировка 'K-A', которая применялась до 1998 года (https://www.playground.ru/misc/opinion/istoriya_rejtinga_esrb_kak_krov_i_nasilie_stali_ne_dlya_detej-714849).  
Заменим ее на актуальную E10+.

```python
data.loc[data['rating'] == 'K-A', 'rating'] = 'E10+'
```

```python
data.query('rating == "K-A"')
```

```python
data['rating'].sort_values(ascending = True).unique()
```

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

```python
# статистика для платформы XOne
actual_platform.loc[actual_platform['platform'] == 'XOne', 'user_score'] .describe()
```

```python
# статистика для платформы PC
actual_platform.loc[actual_platform['platform'] == 'PC', 'user_score'] .describe()
```

*Нулевая гипотеза:*   
Средний пользовательский рейтинг платформы Xbox One равен среднему пользовательскому рейтингу платформы PC.

*Альтернативная гипотеза*.
Средний пользовательский рейтинг платформы Xbox One отличается от среднего пользовательского рейтинга платформы PC.  

Для проверки нулевой гипотез необходимо посчитать р-уровень значимости для статистического различия средних двух генеральных совокупностей по имеющимся у нас выборкам. В соответствии с теорией, поскольку размер выборки значительно больше 30, значимым р-уровнем ( 𝛼 ) можно считать величину 0.05 (5%), поскольку t-распределение средних значений выборок стремиться к нормальному.

```python
alpha = 0.05
result_platform = st.ttest_ind(
    actual_platform.loc[actual_platform['platform'] == 'XOne', 'user_score'].dropna(),
    actual_platform.loc[actual_platform['platform'] == 'PC', 'user_score'].dropna()
)
print('p-значение: {:.2f}'.format(result_platform.pvalue))

if result_platform.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

**Вывод**  
Нулевая гипотеза о равенстве пользовательского рейтинга между платформами Xbox One и PC не может быть отвергнута. Полученное значение p-value говорит о том, что средняя оценка пользователей платформы Xbox One и платформы PC отличается, но с вероятностью в 14% такое различие можно получить случайно. Это слишком высокая вероятность для того, чтобы делать выводы о значимом различии между средними оценка пользователей.

Средний пользовательский рейтинг жанра Action равен среднему пользовательскому рейтингу жанра Sports.

*Альтернативная гипотеза*.
Средний пользовательский рейтинг жанра Action отличается от среднего пользовательского рейтинга жанра Sports. 

Для проверки нулевой гипотез необходимо посчитать р-уровень значимости для статистического различия средних двух генеральных совокупностей по имеющимся у нас выборкам. В соответствии с теорией, поскольку размер выборки значительно больше 30, значимым р-уровнем ( 𝛼 ) можно считать величину 0.05 (5%), поскольку t-распределение средних значений выборок стремиться к нормальному.
